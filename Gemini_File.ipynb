```python
#Importa√ß√µes e configura√ß√µes iniciais
import google.generativeai as genai
from google.colab import userdata
from google.colab import files

api_key = userdata.get('API_KEY')
genai.configure(api_key=api_key)

# Set up the model
generation_config = {
  "temperature": 0,
  "top_k": 1
}

# Configura o modelo Gemini Pro
# Ok - .wav, .mp3, .jpg, .png
# Error - .mp4, .pdf, .docx
model_name='gemini-1.5-pro-latest'

# Configura o modelo Gemini Pro Vision
# Ok - .jpg, .png
# Error - .wav, .pdf, .docx, .mp3, .mp4
# model_name='gemini-1.0-pro-vision-latest'

model = genai.GenerativeModel(model_name=model_name, generation_config=generation_config)
chat = model.start_chat(history=[])

print("Escreva sua pergunta ou comando e em seguida fa√ßa o upload de seu arquivo (opcional):\n")

# Recebe input do usu√°rio (prompt)
prompt = input("USU√ÅRIO:\n")

# Executa o programa at√© o usu√°rio digitar "sair"
while prompt != "sair":
  print("\nClique em \"Choose files\" para fazer upload de um arquivo (.wav, .mp3, .jpg, .png). Caso contr√°rio, clique em \"Cancel upload\" para prosseguir:")
  uploaded = {}
  uploaded = files.upload()

  # Checa se o usu√°rio fez o upload de algum arquivo
  if uploaded != {}:
    # Caso positivo, cria uma lista com o arquivo e o prompt
    for fn in uploaded.keys():
      prompt_parts = [
        genai.upload_file(fn),
        prompt,
      ]
    # Envia requisi√ß√£o com o arquivo para o Gemini e recebe resposta
    response = chat.send_message(prompt_parts)
  else:
    # Caso negativo, envia a requisi√ß√£o sem o arquivo para o Gemini e recebe a resposta
    response = chat.send_message(prompt)

  # Imprime a resposta do Gemini
  print(f"\nGEMINI:\n{response.text}")

  print("-------------------------------------------------------------------------------------------------------------------------\n")

  # Recebe input do usu√°rio (prompt)
  prompt = input("USU√ÅRIO:\n")
```
